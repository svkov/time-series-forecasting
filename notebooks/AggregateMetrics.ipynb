```python
def parse_row(ticker, name, row):
    
    argmin = row.argmin()
    return {
        'metric': name,
        'model': row.index[argmin],
        'value': row[argmin]
    }
```

```python
path = '../reports/metrics/'
input = list(map(lambda x: path + x, os.listdir(path)))
full_df = pd.DataFrame()
for i in input:
    if 'best_metrics' in i:
        continue
    ticker = os.path.split(i)[-1].replace('metrics_', '').replace('.csv', '')
    df = pd.read_csv(i, index_col=0)
    df['ticker'] = ticker
    full_df = full_df.append(df)
full_df['metric'] = full_df.index
full_df = full_df.set_index(['ticker', 'metric'])
full_df
```

```python
def generate_table_header(body, caption):
    if not isinstance(body, str):
        raise ValueError(f'body должен быть строкой, вместо этого {type(body)}')
    return \
        f"""\\begin{{center}}
    \\begin{{longtable}}{{|p{{2cm}}|p{{3cm}}|p{{7cm}}|p{{3cm}}|}}
    \\caption{{{caption}}}
    {body}
\\end{{center}}"""


def generate_row(name, row, sep='\hline\n', end='\\'):
    name = f'{name[0]} & {name[1]} & '
    values = ' & '.join(map(str, row.tolist()))
    return sep + name + values + end


def df_to_latex(df: pd.DataFrame, caption):
    body = ''
    for name, row in df.iterrows():
        row = generate_row(name, row)
        print(row)
        body += row
    return generate_table_header(body, caption)
```
